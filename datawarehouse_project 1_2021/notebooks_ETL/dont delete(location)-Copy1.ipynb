```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('owid-covid-data.xlsx') 

data
```

```python
data.info()
```

```python
dimLocation  = data[["location","iso_code","continent","population","life_expectancy"]]

dimLocation
dimLocation.info()
```

```python
#print(train[train.isnull().any(axis=1)][null_columns].head())
dimLocation[dimLocation["iso_code"].isnull()]
```

```python
dimLocation = dimLocation.dropna()  #drops all null
```

```python
dimLocation= dimLocation.drop_duplicates(subset=['location']) #removes duplicate rows

dimLocation
```

```python
dimLocation["locationID"] = np.arange(1,len(dimLocation)+1)

print(dimLocation[0:1000])
```

```python
dimLocation=dimLocation.reset_index(drop=True) #assigns proper index

dimLocation
```

```python
dimLocation['Size']=np.nan #Adds new empty column
```

```python
#small<=2000000
#large>=40000000

for i in range( len(dimLocation)):
    if(int(dimLocation.population[i])<=2000000.0):
        dimLocation["Size"][i] = "Small"
    elif(int(dimLocation.population[i])>=40000000.0):
        dimLocation["Size"][i] = "Large"
    else:
        dimLocation["Size"][i] = "Medium"
        
dimLocation
```
