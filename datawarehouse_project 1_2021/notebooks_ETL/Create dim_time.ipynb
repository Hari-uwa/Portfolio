```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('time_series_covid19_confirmed_global.csv') #read the csv file

data
```

```python
data=data.drop(['Lat','Long','Province/State','Country/Region'],axis=1) #deletes unwanted columns
```

```python

#dim_time = pd.DataFrame(columns=["Date","Month","dayID"],dtype=[np.datetime64,np.object,np.int64])
dim_time = pd.DataFrame(columns=["Date","Month","Year","dayID"]) #creates a new dataframe and adds columns
```

```python
data_T=data.T #Transpose data

data_T
```

```python
dim_time["Date"]=data_T.index #Adds value to column "date"
#dim_time["Date"]=dim_time["Date"].astype('datetime64[ns]')
```

```python
#forloop to add values under 'months' and 'year'

months = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June","7":"July","8":"August","9":"September","10":"October","11":"November","12":"December"}
for i in range(len(dim_time)):
    date_value = dim_time.Date[i]
    for key in months:
        if (key == str(date_value[:date_value.find('/')])):
            dim_time.Month[i] = months[key]
            if(str(date_value[-2:])=="20"):
                dim_time.Year[i] = 2020
            else:
                dim_time.Year[i] = 2021
                
dim_time.tail()
```

```python
dim_time["dayID"] = np.arange(1,len(dim_time)+1) #Assigns a dayID(for a surrogate key)
```

```python
dim_time
```

```python
dim_time["Date"]=dim_time["Date"].astype('datetime64[ns]') #Makes the date column to date-time format
```
